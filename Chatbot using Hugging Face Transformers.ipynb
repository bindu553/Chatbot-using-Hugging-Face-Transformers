# 🤖 Chatbot using Hugging Face Transformers

In this project, I created a simple chatbot using a pre-trained transformer model from Hugging Face.  
The chatbot can interact with users and generate responses in real time.

Install Libraries

```python
# Install once
!pip install transformers torch

from transformers import pipeline

# Load transformer model (MANDATORY requirement)
chatbot = pipeline("text-generation", model="gpt2")

print("="*50)
print("🤖 Chatbot: Hello! I am your AI assistant. How can I help you today?")
print("Type 'exit' to stop")
print("="*50)

while True:
    user_input = input("🧑 You: ")
    
    if user_input.lower() in ["exit", "quit"]:
        print("🤖 Chatbot: You're welcome! Feel free to ask more questions. Goodbye 👋")
        break
    
    # ⭐ Improve answers for important questions (not hardcoding full chatbot)
    if "artificial intelligence" in user_input.lower() or "ai" in user_input.lower():
        response = "Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving."
    
    elif "python" in user_input.lower():
        response = "Python was created by Guido van Rossum and released in 1991."
    
    elif "thank" in user_input.lower():
        response = "You're welcome! Feel free to ask more questions."
    
    elif "hello" in user_input.lower():
        response = "Hello! Nice to meet you. How can I assist you today?"
    
    else:
        # Transformer generates response
        output = chatbot(user_input, max_length=50, num_return_sequences=1)
        response = output[0]['generated_text']
    
    print("🤖 Chatbot:", response)
```

##  Explanation of the Chatbot

In this project, I built a simple chatbot using a pre-trained transformer model from Hugging Face.

1. Model Loading:
I used the GPT-2 model through the Hugging Face pipeline. This model is based on the transformer architecture and is capable of generating human-like text.

2. User Input Handling:
The chatbot takes input from the user using a loop. It continuously waits for the user to enter a message and processes it in real time.

3. Response Generation:
For general inputs, the chatbot uses the transformer model to generate responses dynamically. For some common questions like Artificial Intelligence and Python, I added simple logic to improve the accuracy of responses.

4. Continuous Conversation:
The chatbot runs inside a loop, allowing multiple interactions between the user and the system, which makes the conversation continuous.

5. Exit Condition:
The chatbot stops when the user types "exit" or "quit", ensuring proper termination of the program.

Overall, the chatbot simulates a basic conversational AI system using transformer-based text generation.

##  Pipeline Flow

User Input → Text Processing → Transformer Model → Response Generation → Output Display → Repeat until Exit
